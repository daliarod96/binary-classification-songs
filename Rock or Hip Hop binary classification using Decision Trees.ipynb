```python
clean_data = data.copy()
clean_data['isRock'] = (clean_data['genre_top'] == 'Rock')*1
print(clean_data['isRock'])
```

```python
y=clean_data[['isRock']]
```

```python
clean_data['genre_top']
```

```python
y.head()
```

```python
genre_features = ['bit_rate', 'duration', 'listens']
```

```python
X = clean_data[genre_features].copy()
```

```python
X.columns
```

```python
y.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.33, random_state=324)
```

```python
rock_classifier = DecisionTreeClassifier(max_leaf_nodes = 10, random_state = 0)
rock_classifier.fit(X_train, y_train)
```

```python
type(rock_classifier)
```

```python
predictions = rock_classifier.predict(X_test)
```

```python
predictions[:10]
```

```python
y_test['isRock'][:10]
```

```python
accuracy_score(y_true = y_test, y_pred = predictions)
```
